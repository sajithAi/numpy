```python
import numpy as np
# CREATE  a 4x4 martix filled with zeros
arr=np.zeros((4,4))
print(arr)
print("--------------------------------")
# fill it with checkboard pattern
arr[1::2,::2]=1
# 1::2, selects every second row, starting from the second row (index 1).
# ::2 selects every second column, starting from the first column(index 0 )
print(arr)
arr[::2,1::2]=1
# ::2 selects every second row, starting from the first row (index 0)
# 1::2, selects every second column, starting from the first row (index 1)
print("=========final result===========")
print(arr)
```

# Numpy array copy and vs view

## The difference between copy and view



*   Copy is a new array, and the view is just a view of the original array.
*   copy owns the data any changes made to the copy will not affect original array,
*  the views does not own the data and any changes made to the view will affect the original array.



```python
import numpy as np
arr=np.array([1,2,3,4,5])
x=arr.copy()
arr[0]=42
print(arr) # display arr after changing the first element to 42
print(x)   # copy of arr is store

```

# JOINING ARRAY USING STACK FUNCTION:

* stacking is same as concatenation,the only difference is that stacking is done along a new axis.

* we can concatenate two 1-d arrays along the second axis which would result in putting them one over the other.

* we pass a sequence of arrays that we want to join ti the stack()method along with the axis.if axis is not explicitly passed it is taken as 0,

```python
import numpy as np
a1=np.array([1,2,3])
a2=np.array([4,5,6])
arr=np.stack((a1,a2),axis=1)
print(arr)
```

```python
import numpy as np
a1=np.array([1,2,3])
a2=np.array([4,5,6])
print(a1)
print("\n",a2)
arr_stack1=np.stack((a1,a2),axis=1)
print(arr_stack1)
arr_stack0=np.stack((a1,a2),axis=0)
print(arr_stack0)
arr_con=np.concatenate((a1,a2))
print(arr_con)

```

# NUMPY SPLITTING ARRAY

* splitting is reverse operation of joining.

* splitting breaks one array into multiple.

* we use array_split() for splitting arrays,we pass it the array we want to split and the number of splits.

: SPLIT THE ARRAY IN 3 PARTS:

```python
import numpy as np
a=np.array([1,2,3,4,5,6])
newarr=np.array_split(a,3)
print(newarr)
```

: SPLIT INTO ARRAYS:

* the return value of the array_split() method is an array containing each of the split as an array.

* if you an array into 3 arrays, you can access them from the result just like any array element.

```python
import numpy as np
a=np.array([1,2,3,4,5,6])
newarr=np.array_split(a,3)
print(newarr[0])
print(newarr[1])
print(newarr[2])
print(*newarr)
```

Q. Create a filter array that will return only values higher than 42:

```python
import numpy as np
a=np.array([41,42,43,44])
filter_arr=[]
for element in a:
  if element > 42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=a[filter_arr]
print("original array:\n",a)
print("\nboolean array:\n",filter_arr)
print("\nfilter new array:\n",newarr)
```

The Zipf distribution, also known as Zipf's law, is a discrete probability distribution that follows a power-law pattern. #  Create Your Own ufunc

## How To Create Your Own ufunc
• To create your own ufunc, you have to define a function, like you do with normal functions in Python, then you add it to your NumPy ufunc library with the frompyfunc() method.

## The frompyfunc() method takes the following arguments:

* function-the name of the function.
* inputs- the number of input arguments (arrays).
* outputs- the number of output arrays.


# Create your own ufunc for addition:

```python
import numpy as np
def myadd(x,y):
  return x+y
myadd=np.frompyfunc(myadd,2,1)
print(myadd([1,2,3,4],[5,6,7,8]))
```

```python
import numpy as np
def mymultiplication(x,y):
  return x*y
mymultiplication=np.frompyfunc(mymultiplication,2,1)
print(mymultiplication([1,2,3,4],[5,6,7,8]))
```

### To test if the function is a ufunc in an if statement, use the numpy.ufunc value (or np.ufunc if you use np as an alias for numpy):

```python
# use an if statement to check if the function is a ufunc or not:
import numpy as np
if type(np.add)==np.ufunc:
  print('add is ufunc')
else:
  print('add is not ufunc')
```

#  Simple Arithmetic

 ## Simple Arithmetic
 * You could use arithmetic operators +-*/ directly between NumPy arrays, but this section discusses an extension of the same where we have functions that can take any array-like objects e.g. lists, tuples etc. and perform arithmetic conditionally.
* Arithmetic Conditionally: means that we can define conditions where the arithmetic operation should happen
*  All of the discussed arithmetic functions take a where parameter in which we can specify that condition.

# Addition
* The add() function sums the content of two arrays, and return the results in a new array.

```python
# add the values in arr1 to the values in arr2:
import numpy as np
arr1=np.array([10,11,12,13,14,15])
arr2=np.array([20,21,22,23,24,25])
newarr=np.add(arr1,arr2)
print(newarr)
```

# subtraction

 * the subtract() function subtracts the values from another array and return the results in a new array

# subtract the values in arr2 from the values in arr1:


```python
import numpy as np

arr1=np.array([10,20,30,40,50,60])
arr2=np.array([20,21,22,23,24,25])
newarr=np.subtract(arr1,arr2)
print(newarr)
```

# multiplication

* The multiply() function multiplies the values from one array with the values from another array, and return the results in a new array.

```python
# multiplication the values in arr1 with the values in arr2:
import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([20,21,22,23,24,25])
newarr=np.multiply(arr1,arr2)
print(newarr)
```

# Division

* The divide() function divides the values from one array with the values from another array, and return the results in a new array.


```python
# divide the values in arr1 with the values in arr2:
import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([3,5,10,8,2,33])
newarr=np.divide(arr1,arr2)
print(newarr)
```

# Power

* The power() function rises the values from the first array to the power of the values of the second array, and return the results in a new array.


```python
# raise the values in arr1 to the power of values in arr2:
import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([3,5,6,8,2,33])
newarr=np.power(arr1,arr2)
print(newarr)
```

# Remainder


Both the mod() and the remainder() functions return the remainder of the values in the first array corresponding to the values in the second array, and return the results in a new array.

```python
# both the remainders:
import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([3,7,9,8,2,33])
newarr1=np.mod(arr1,arr2)
print(newarr1)
```

```python
# you get the same result when using the remainder() function:
import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([3,7,9,8,2,33])
newarr2=np.remainder(arr1,arr2)
print(newarr2)
```

# **Quotient and mod**
* the divmod()function return both the quotient and the mod .the return values is two arrays ,the arrays, the first array contains the quotients and second array contains the mod

```python
import numpy as np
arr1=np.array([10,20,30,40,50,60])
arr2=np.array([3,7,9,8,2,33])
newarr=np.divmod(arr1,arr2)
print(newarr)
```
